```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression as lr 
from sklearn.ensemble import RandomForestClassifier as rfc 
from sklearn.svm import LinearSVC as lsvc 
```

```python
#read the winner and losers files into 2 seperate dataframes for winners and losers

wPATH = "https://raw.githubusercontent.com/Canadiak/Deerhacks-Web-Scraping/master/csv_folder/winner_{0}.txt"
lPATH = "https://raw.githubusercontent.com/Canadiak/Deerhacks-Web-Scraping/master/csv_folder/loser_{0}.txt"
winners = pd.read_csv(wPATH.format(1990), usecols=range(53))
losers  = pd.read_csv(lPATH.format(1990), usecols=range(53))
for i in range(1991, 2020):
  winners = winners.append(pd.read_csv(wPATH.format(i), usecols=range(53)), ignore_index=True)
  losers = losers.append(pd.read_csv(lPATH.format(i), usecols=range(53)), ignore_index=True)
```

```python
winners
```

```python
losers
```

```python
#shuffle both datasets using the same indexing so they stay in the same
#relative order. Doing it this way ensures that there's a balance between
#the age of data being assigned to zeros and ones later

idx = np.random.permutation(winners.index)
winners = winners.reindex(idx)
losers = losers.reindex(idx)
```

```python
#get rid of all null columns
winners = winners.dropna(axis=1)
losers  = losers.dropna(axis=1)
```

```python
#get rid of all non numeric columns
winners = winners._get_numeric_data()
losers  = losers._get_numeric_data()
```

```python
#split data in half into zeros and ones. zeros are where the 
#winning team is logged first and ones are where the losing 
#team is logged first.

zeros = winners.iloc[:225, :].sub(losers.iloc[:225, :])
ones  = losers.iloc[225:, :].sub(winners.iloc[225:, :])
```

```python
zeros['winner'] = 0
zeros
```

```python
ones['winner'] = 1
ones
```

```python
#combining both datasets into 1

data = pd.concat([zeros, ones])
```

```python
data.columns
```

```python
#dropping columns that I already know are useless or redundant 

data = data.drop(['Rk', 'G', 'MP', 'FG', '3P', '2P', 'FT', 'TRB', 'PTS', 'Team2', 'L', 'PW', 'MOV', 'SOS', 'SRS', 'NRtg', '3PAr', 'Arena', 'eFG%', 'TOV%', 'ORB%', 'FT/FGA', 'TOV%2', 'DRB%', 'FT/FGA2'], axis=1)
data
```

```python
data.columns
```

```python
from sklearn.utils import shuffle 
data = shuffle(data)
```

```python
y = data['winner']
y
```

```python
X = data.drop(['winner'], axis=1)
X
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler as ss

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42, shuffle=True)
```

```python
model1 = make_pipeline(ss(), lr())
model1.fit(X_train, y_train)
```

```python
model1.score(X_test, y_test)
```

```python
#80% is pretty good but i'm gonna see if i can do better
from sklearn.feature_selection import SelectKBest, f_classif

X_new = SelectKBest(f_classif, k=17).fit(X_train, y_train)
```

```python
X_new.transform(X_train)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42, shuffle=True)
model2 = make_pipeline(ss(), lr())
model2.fit(X_new.transform(X_train), y_train)
model2.score(X_new.transform(X_test), y_test)
```
